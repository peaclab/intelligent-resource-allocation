```python
# Prepare the data for plotting
datasets = list(results_dict.keys())
rmse_user = [results_dict[dataset]['rmse_user'] for dataset in datasets]
mae_user = [results_dict[dataset]['mae_user'] for dataset in datasets]
r2_user = [results_dict[dataset]['r2_user'] for dataset in datasets]

rmse_single = [results_dict[dataset]['rmse_single'] for dataset in datasets]
mae_single = [results_dict[dataset]['mae_single'] for dataset in datasets]
r2_single = [results_dict[dataset]['r2_single'] for dataset in datasets]

rmse_cluster = [results_dict[dataset]['rmse_cluster'] for dataset in datasets]
mae_cluster = [results_dict[dataset]['mae_cluster'] for dataset in datasets]
r2_cluster = [results_dict[dataset]['r2_cluster'] for dataset in datasets]

# Plotting RMSE, MAE, and R² for each dataset
x = np.arange(len(datasets))  # the label locations
width = 0.2  # the width of the bars

fig, ax = plt.subplots(1, 3, figsize=(18, 6))

# RMSE Plot
ax[0].bar(x - width, rmse_user, width, label='User Requests', color='b')
ax[0].bar(x, rmse_single, width, label='Single Model Prediction', color='g')
ax[0].bar(x + width, rmse_cluster, width, label='Multiple Model Prediction (Avg)', color='r')
ax[0].set_ylabel('RMSE')
ax[0].set_title('RMSE Comparison')
ax[0].set_xticks(x)
ax[0].set_xticklabels(datasets)
ax[0].legend()

# MAE Plot
ax[1].bar(x - width, mae_user, width, label='User Requests', color='b')
ax[1].bar(x, mae_single, width, label='Single Model Prediction', color='g')
ax[1].bar(x + width, mae_cluster, width, label='Multiple Model Prediction (Avg)', color='r')
ax[1].set_ylabel('MAE')
ax[1].set_title('MAE Comparison')
ax[1].set_xticks(x)
ax[1].set_xticklabels(datasets)
ax[1].legend()

# R² Plot
ax[2].bar(x - width/2, r2_single, width, label='Single Model Prediction', color='b')
ax[2].bar(x + width/2, r2_cluster, width, label='Multiple Model Prediction (Avg)', color='r')
ax[2].set_ylabel('R²')
ax[2].set_title('R² Comparison')
ax[2].set_xticks(x)
ax[2].set_xticklabels(datasets)
ax[2].legend()

fig.suptitle('Execution Time Prediction Performance Metrics Comparison Across Datasets', fontsize=16, fontweight='bold')

fig.tight_layout(rect=[0, 0, 1, 0.95])  # Leave space for the suptitle
plt.show()
```
